```python
#Exercise 1
import math
import numpy as np
x = np.random.choice ([0, 1, 2, 3, 4, 5], 3650, p=[0.1, 0.2, 0.3, 0.2, 0.15, 0.05])
x = x.tolist()
#a
X = set(x)
print(X)

#b
P = [x.count(i)/len(x) for i in X]
print(P)

#c
EX = 0
for x in X:
  EX = EX+ (x * P[x-1])
print ("expectation : ", EX)

VarX = 0
for x in X:
  VarX = VarX + (x-EX)*(x-EX)*P[x-1]
print("variance : ", VarX)

SD = math.sqrt(VarX)
print("standard deviation : ", SD)

#d
FX = sum(P[3:])
print("the probability of having 3 or more emergency cases : ",FX)
```
